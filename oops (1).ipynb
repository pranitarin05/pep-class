```python
#inheritance
#one class user another class properties fucntion
class human:
  def eat(self):
    print("human can eat")
class student(human):
  def study(self):
    print("student can study ")
s1=student()
s1.eat()
s1.study()
```

```python
#inheritance code of animal can speak and dog can bark
class animal:
  def speak(self):
    print("animal can speak")
class dog(animal):
  def bark(self):
    print("dogs bark")
d1=dog()
d1.speak()
d1.bark()
```
